```python
from typing import List, Tuple
import rosbag
from os import path
import numpy as np

from utils import create_scattermapbox, plot

GPS_TOPIC = '/gps/gps'


def read_rosbag_gps(bag_path: str, gps_topic: str) -> np.array:
    points = []
    print(f'reading {bag_path}')
    with rosbag.Bag(bag_path, 'r') as bag:
        messages = bag.read_messages(topics=[gps_topic])
        for topic, msg, timestamp in messages:
            points.append((msg.latitude, msg.longitude, msg.altitude))
    
    return np.array(points)


data = {
    'vabadusepst': read_rosbag_gps(path.expanduser('~/bags/vabadusepst.bag'), GPS_TOPIC),
    'vabadussild': read_rosbag_gps(path.expanduser('~/bags/vabadussild.bag'), GPS_TOPIC),
    'oldtown': read_rosbag_gps(path.expanduser('~/bags/oldtown.bag'), GPS_TOPIC),
}
```

```python

traces = []
for name, trajectory in data.items():
    traces.append(create_scattermapbox(trajectory, name, mode='lines', bold=True))

tmp = np.concatenate([*data.values()])
lat_center = np.mean(tmp[:,0])
alt_center = np.mean(tmp[:,1])

plot(traces, (lat_center, alt_center))
```
