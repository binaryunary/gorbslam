```python
from os import path

import plotly.graph_objects as go

from utils import read_tum_file


# TRAJECTORY_NAME = 'm_vabadusepst'
TRAJECTORY_NAME = 'm_oldtown'

trajectories = []
for i in range(0, 5):
  trajectories.append(read_tum_file(path.join('data', TRAJECTORY_NAME, f'm_trajectory_{i}.txt')))

fig = go.Figure()

for i, trajectory in enumerate(trajectories):
  fig.add_trace(go.Scatter(x=trajectory[:, 1], y=trajectory[:, 3], mode='lines', name=f'kf_trajectory_{i}'))  

fig.update_layout(yaxis=dict(scaleanchor="x", scaleratio=1), height=800)
fig.show()
```
