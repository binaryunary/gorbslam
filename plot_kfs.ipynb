```python
from os import path

import plotly.graph_objects as go

from orbslam_results import read_kf_trajectory

trajectories = []
for i in range(0, 4):
  trajectories.append(read_kf_trajectory(path.join('data', 'vabadusepst_kf', f'kf_trajectory_{i}.txt')))

fig = go.Figure()

for i, trajectory in enumerate(trajectories):
  fig.add_trace(go.Scatter(x=trajectory[:, 0], y=trajectory[:, 2], mode='lines', name=f'kf_trajectory_{i}'))  

fig.update_layout(yaxis=dict(scaleanchor="x", scaleratio=1), height=800)
fig.show()
```
